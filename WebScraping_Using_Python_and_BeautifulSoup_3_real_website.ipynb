# Scarping on real website using request library

```python
import requests
from bs4 import BeautifulSoup
```

```python
# Aim is to get all the jobs posted few days ago
# Intrested in getting the text 
html_text = requests.get("https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=python&txtLocation=").text
```

```python
print(html_text)
```

```python
# Creating soup
soup = BeautifulSoup(html_text,"lxml")
jobs = soup.find_all('li', class_= "clearfix job-bx wht-shd-bx")
# As we can see this page having many of job lists, but it is going to work or the particular present page
print(jobs)
```

```python
# printing the companies name present in the page
company_names = soup.find_all('h3',class_= "joblist-comp-name")
for name in company_names:
    print(name.text)
```

```python
# Rather than finding company name in entire page, we are just intrested in getting 
# company name in jobs
for job in jobs:
    company_name = job.find_all('h3', class_ = "joblist-comp-name")
    for name in company_name:
        print(name.text.replace(" ",""))   # as we are seeing in company names, we are getting extra white spaces, we replace it with nothing
```

```python
# Let's find out the skills
for job in jobs:
    skills = job.find_all('span', class_ = "srp-skills")
    for skill in skills:
        print(skill.text.replace(" ","")) #.replace to replace white spaces with nothing
```

# Storing the results in a file

```python

print("Put the unfamiliar skills")
unfamiliar_skills = input()  # if this skill are having in skill set we just ignore that job post
print(f"Filter_out_unfamiliar_job: {unfamiliar_skills}")

def find_job():
    html_text = requests.get("https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=python&txtLocation=").text
    soup = BeautifulSoup(html_text,"lxml")
    jobs = soup.find_all('li', class_= "clearfix job-bx wht-shd-bx")
    for index,job in enumerate(jobs):   # enumerate iterate on all the index through their indexes
        published_date = job.find('span', class_="sim-posted").span.text
        if "few" in published_date :
            company_name = job.find('h3', class_ = "joblist-comp-name").text.replace(" ","")
            skill = job.find('span', class_ = "srp-skills").text.replace(" ","")
            more_info = job.header.h2.a['href']
            if unfamiliar_skills not in skill:
                with open(f'{index}.txt', 'w') as f:
                    f.write(f"Company name : {company_name.strip()} \n")
                    f.write(f"Skills: {skill.strip()} \n")
                    f.write(f"More_info: {more_info} ")
            print(f"File Saved: {index}")

                
if __name__ == "__main__":
    while True:
        find_job()
#         time_wait = 2
#         print(f"waitint {time_wait} minutes......")
#         time.sleep(time_wait*60) 
  # allow program toi wait for certain amount of time and it is in seconds
```
